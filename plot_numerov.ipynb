```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data
data = np.genfromtxt('y.dat', delimiter=' ')
```

```python
# Plot the data 
fig = plt.figure(dpi=120)
plt.plot(data[:,0], data[:,1], color='k')
plt.xlabel('x')
plt.ylabel('y')
plt.title('F(x) = -x')
# plt.savefig('numerov_plot_minusx.png')
plt.show()
```
